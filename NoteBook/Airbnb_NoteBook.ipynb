```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("../CSV/AB_NYC_2019.csv")
df= pd.DataFrame(df)
```

```python
df.info()
df.describe()
df.nunique()
df.isna().sum()
```

Data Cleaning 

```python
df[["name","host_name"]]= df[["name","host_name"]].fillna("Unknown")
df["last_review"] = pd.to_datetime(df["last_review"])
df["reviews_per_month"] = df["reviews_per_month"].fillna(0,inplace=True)
df.drop("reviews_per_month",axis=1, inplace=True)
```

Feature Engineering

```python
df["has_reviews"] = (df["number_of_reviews"]> 0).astype(int)
```

```python
df["review_lavel"] = pd.cut(df["number_of_reviews"],
                           bins=[-1,0,10,50,200,1000],
                           labels =["No Review","Low","Medium", "High","Very High"])
```

```python
df["price_category"] = pd.cut(df["price"],
                             bins= [-1,50,150,300,1000,10000],
                             labels = ["Budget", "Standard", "Premium", "Luxury","Ultra Luxury"])
```

```python
df["stay_type"] = df["minimum_nights"].apply(
                                        lambda x:"Short" if x<=3 else"Medium" if x<=7 else "Long")
```

```python
df["availability_level"] = pd.cut(df["availability_365"],
                                 bins = [-1,100,200,365],
                                 labels = ["Low","Medium", "High"])
```

```python
df["host_type"] = df["calculated_host_listings_count"].apply(
lambda x: "Single" if x == 1 else "Professional" )
```

```python
df.to_csv("cleaned_airbnb.csv", index= False)
```
