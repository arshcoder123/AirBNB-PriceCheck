```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../CSV/cleaned_airbnb.csv")
df = pd.DataFrame(df)
df["last_review"] = pd.to_datetime(df["last_review"])
```

```python
df_filtered = df[df["price"]<1000]
plt.hist(df_filtered["price"],bins=5)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count of Room")
plt.show()
```

This indicates that the Airbnb market is dominated by budget and mid-range properties, with luxury listings forming only a small fraction of the total supply.

```python
df.boxplot(column= "price", by= "room_type")
plt.title("Price Vs Room Types")
plt.xlabel("Room Types")
plt.ylabel("Price in ($)")
plt.show()
```

The pricing structure clearly aligns with customer preferences for privacy, indicating that users are willing to pay a premium for entire properties compared to shared accommodations.

```python
dfloc = df.groupby("neighbourhood_group")["price"].mean().plot(kind="bar")
plt.title("Price Vs Location")
plt.xlabel("Location")
plt.ylabel("Price In ($)")
plt.show()
```

This indicates that location plays a major role in pricing, with central and high-demand areas like Manhattan commanding premium prices due to better accessibility and tourist demand.

```python
plt.scatter(df["number_of_reviews"],df["price"])
plt.xlabel("Number Of Reviews")
plt.ylabel("Price In ($)")
plt.show()
```

This suggests that more affordable listings attract higher booking volumes, leading to more reviews, whereas expensive listings cater to a niche audience with lower booking frequency.

```python
plt.scatter(df["availability_365"],df["price"])
plt.title("Availability Vs Price")
plt.xlabel("Availability in Days")
plt.ylabel("Price In ($)")
plt.show()
```

Listings with higher availability tend to have lower prices, suggesting that less frequently booked properties reduce prices to attract customers.
